**PRIMER GRÁFICO**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("survey_data_updated.csv")
print("Columnas cargadas:")
print(df.columns)
```

```python
print("\nEjemplo de LanguageHaveWorkedWith:")
print(df["LanguageHaveWorkedWith"].head())
```

```python
df_lang = df["LanguageHaveWorkedWith"].dropna()

# Separar cada fila "Python;Go;Rust" → en múltiples
all_langs = df_lang.str.split(";").explode()

all_langs = all_langs.str.strip()

lang_counts = all_langs.value_counts()

print("\nConteo de lenguajes (Top 20):")
print(lang_counts.head(20))
```

```python
top10 = lang_counts.head(10)
print("\nTOP 10 lenguajes más usados:")
print(top10)
```

```python
top10.to_csv("top10_languages.csv", header=["Count"])
print("\nArchivo generado: top10_languages.csv")
```

**SEGUNDO GRÁFICO**

```python
# Asegurar que no haya valores nulos
df['DatabaseHaveWorkedWith'] = df['DatabaseHaveWorkedWith'].fillna("")

# Separar los valores múltiples
df_exploded = (
    df.assign(DatabaseHaveWorkedWith = df['DatabaseHaveWorkedWith'].str.split(';'))
      .explode('DatabaseHaveWorkedWith')
)

# Eliminar espacios sobrantes
df_exploded['DatabaseHaveWorkedWith'] = df_exploded['DatabaseHaveWorkedWith'].str.strip()

# Eliminar vacíos
df_exploded = df_exploded[df_exploded['DatabaseHaveWorkedWith'] != ""]

# Contar apariciones
db_counts = df_exploded['DatabaseHaveWorkedWith'].value_counts().head(10)

# Convertir a DataFrame
df_db_top10 = db_counts.reset_index()
df_db_top10.columns = ['database', 'count']

# Guardar archivo para Cognos
df_db_top10.to_csv("top10_databases.csv", index=False)

df_db_top10
```
