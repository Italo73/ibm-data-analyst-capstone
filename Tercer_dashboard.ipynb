```python
import pandas as pd

df = pd.read_csv("survey_data_updated.csv")
df.head()
```

```python
# Quitar nulos
df_age = df['Age'].fillna("No especificado")

# Contar ocurrencias
age_counts = df_age.value_counts().reset_index()
age_counts.columns = ["Age", "count"]

# Guardar CSV
age_counts.to_csv("demografia_age_distribution.csv", index=False)

age_counts
```

```python
# Llenar vacíos
df_country = df['Country'].fillna("No especificado")

country_counts = df_country.value_counts().reset_index()
country_counts.columns = ["Country", "count"]

country_counts.to_csv("demografia_country_counts.csv", index=False)

country_counts
```

```python
df_ed = df['EdLevel'].fillna("No especificado")

education_counts = df_ed.value_counts().reset_index()
education_counts.columns = ["EdLevel", "count"]

education_counts.to_csv("demografia_education_distribution.csv", index=False)

education_counts
```

```python
df_age_ed = df[['Age', 'EdLevel']].copy()

df_age_ed['Age'] = df_age_ed['Age'].fillna("No especificado")
df_age_ed['EdLevel'] = df_age_ed['EdLevel'].fillna("No especificado")

# Conteo combinado
age_ed_counts = (
    df_age_ed
    .groupby(['Age', 'EdLevel'])
    .size()
    .reset_index(name='count')
)

age_ed_counts.to_csv("demografia_age_by_education.csv", index=False)

age_ed_counts
```
