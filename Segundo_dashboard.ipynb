```python
import pandas as pd

# Cargar el archivo actualizado
df = pd.read_csv("survey_data_updated.csv")

df.head()
```

```python
def top10_multivalue(df, column_name, output_filename):
    """
    Procesa una columna multivaluada separada por ';' y genera un Top 10.
    Guarda un CSV listo para usar en IBM Cognos.
    """

    # Asegurar que no haya valores nulos
    df[column_name] = df[column_name].fillna("")

    # Separar valores múltiples
    df_exploded = (
        df.assign(**{column_name: df[column_name].str.split(';')})
          .explode(column_name)
    )

    # Eliminar espacios
    df_exploded[column_name] = df_exploded[column_name].str.strip()

    # Quitar vacíos
    df_exploded = df_exploded[df_exploded[column_name] != ""]

    # Contar ocurrencias
    counts = df_exploded[column_name].value_counts().head(10)

    # Convertir a DF
    result = counts.reset_index()
    result.columns = [column_name, "count"]

    # Guardar CSV para Cognos
    result.to_csv(output_filename, index=False)

    return result
```

```python
df_top10_lang_want = top10_multivalue(
    df,
    "LanguageWantToWorkWith",
    "top10_LanguageWantToWorkWith.csv"
)

df_top10_lang_want
```
